```python
def changeInLinesToZero(chessboard, queenCoordinate):
    chessboard[:, queenCoordinate[1]] = 0
    chessboard[queenCoordinate[0]:queenCoordinate[0]+1, :] = 0
```

```python
def changeDiagonalsToZero(chessboard, queenCoordinate):
    
    #upperleftQuadrant
    y = queenCoordinate[0]
    x = queenCoordinate[1]
    while(y >= 0 and x >= 0):
        chessboard[y,x] = 0
        y = y - 1
        x = x - 1
    #UpperRightQuadrant
    y = queenCoordinate[0]
    x = queenCoordinate[1]
    while(y >= 0 and x < size):
        chessboard[y,x] = 0
        y = y - 1
        x = x + 1
    #LowerLeftQuadrant
    y = queenCoordinate[0]
    x = queenCoordinate[1]
    while(y < size and x >= 0):
        chessboard[y,x] = 0
        y = y + 1
        x = x - 1
    #LowerRightQuadrant
    y = queenCoordinate[0]
    x = queenCoordinate[1]
    while(y < size and x < size):
        chessboard[y,x] = 0
        y = y + 1
        x = x + 1
    
        
```

```python
###Start at 0,0
# add queen and flip zeros
#add queen to next open spot and flip to zeros
#repeat previous step until there is no spots left
#Revert previous step and add queen to spot after next spot
import random as rand
def findNewQueenCoordinate(chessboard):
    yValue = rand.randint(0, size-1)
    xValue = rand.randint(0, size-1)
    if chessboard[yValue][xValue] == 1:
        return (yValue, xValue)
    return (-1, -1)
```
